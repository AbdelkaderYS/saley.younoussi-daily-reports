```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Exercise 1: Create a simple line plot for y = x for x values ranging from 0 to 10.
x = np.linspace(0,10,100)
y = x
plt.figure(figsize=(8,5))
plt.plot(x,y)
plt.show()
```

```python
#Exercise 2: Plot y = x^2 using Matplotlib for x values between -10 and 10.
x = np.linspace(-10,10,100)
y = x**2
plt.figure(figsize=(8,5))
plt.plot(x,y)
plt.show()
```

```python
#Exercise 3: Create a scatter plot with 50 random points where x and y values are between 0 and 10.
plt.figure(figsize=(8,5))
x = np.random.rand(50)*10
y = np.random.rand(50)*10
plt.scatter(x,y)
plt.show()
```

```python
#Exercise4: Plot a sine wave using Matplotlib for x values between 0 and 2.
x = np.linspace(0, 2*np.pi ,100)
y = np.sin(x)
plt.figure(figsize=(8,5))
plt.plot(x,y)
plt.show()
```

```python
# Exercise 5: Create a bar chart for the following data: categories = ['A', 'B', 'C', 'D'] and values = [3, 7, 5, 9].
categories, values = ['A', 'B', 'C', 'D'], [3, 7, 5, 9]
plt.figure(figsize=(8,5))
plt.bar(categories, values)
plt.show()
```

```python
#Exercise 7: Create a histogram of 1000 random numbers generated from a normal distribution.
x = np.linspace(0, 10 ,100)
y1 = x**2
y2 = x**3
plt.figure(figsize=(8,5))
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
#Exercise 8: Add titles, axis labels, and a legend to the plot in Exercise 7.
x = np.linspace(0, 10 ,100)
y1 = x**2
y2 = x**3
#plt.figure(figsize=(6,4))
plt.plot(x,y1, label ="$x^2$")
plt.plot(x,y2, label ="$x^3$")
plt.title(r"$y_1 = x^2$ and $y_2 = x^3 $ functions")
plt.legend()
plt.show()
```

```python
#Exercise 9: Create a subplot with 2 rows and 1 column. Plot y1 = x^2 in the first subplot and y2 = x^3 in the second.
```

```python
x = np.linspace(0, 10 ,100)
y1 = x**2
y2 = x**3
#plt.figure(figsize=(8,4))

plt.subplot(2,1,1)
plt.plot(x,y1, label ="$x^2$", color = "r")
plt.legend()

#plt.title(r"$y_1 = x^2$ and $y_2 = x^3 $ functions")

plt.subplot(2,1,2)
plt.plot(x,y2, label ="$x^3$", color = "b")
plt.legend()
plt.show()
```

```python
# Exercise 10: Use a different color and linestyle for each line in a plot of three functions: y1 = x, y2 = x^2, and y3 = x^3.

x = np.linspace(0, 10 ,100)
y1 = x
y2 = x**2
y3 = x**3
#plt.figure(figsize=(6,4))
plt.plot(x,y1, "r-", label ="$x$")
plt.plot(x,y2,"g*", label ="$x^2$")
plt.plot(x,y3, "b--", label ="$x^3$")
plt.title(r"$y_1 = x$, $y_2 = x^2$ and $y_3 = x^3 $ functions")
plt.legend()
plt.show()
```

```python
# Exercise 20: Sine wave (dashed) and cosine wave (solid) in one plot
x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.plot(x, y_sin, linestyle='--', label='Sine')
plt.plot(x, y_cos, linestyle='-', label='Cosine')
plt.legend()
plt.title("Sine (dashed) and Cosine (solid) waves")
plt.show()
```

```python
# Exercise 21: Contour plot of z = sin(x) + cos(y)
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) + np.cos(Y)
plt.contour(X, Y, Z)
plt.title("Contour plot of z = sin(x) + cos(y)")
plt.show()
```

```python
# Exercise 26: 3D plot of z = x^2 + y^2
from mpl_toolkits.mplot3d import Axes3D
#fig = plt.figure(figsize=(10, 6), dpi=80)
#ax = Axes3D(fig)
fig = plt.figure(figsize=(10,4))
ax = fig.add_subplot(111, projection='3d')
X, Y = np.meshgrid(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100))
Z = X**2 + Y**2
ax.plot_surface(X, Y, Z, cmap="viridis")
plt.title(r"3D plot of z = $x^2 + y^2$")
plt.show()
```
